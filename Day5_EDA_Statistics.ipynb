```python
# Importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
# Load tips dataset
tips = sns.load_dataset('tips')
```

```python
# Basic EDA
print(tips.describe())
sns.histplot(tips['total_bill'], kde=True)
plt.title('Distribution of Total Bill')
plt.show()
```

### Commentary:
The distribution of total_bill is right-skewed with most values between $10 and $20. 
A few high-ticket bills pull the mean above the median. 
This has implications for pricing strategy: consider introducing mid-tier items to normalize revenue spread.

```python
# Sampling + Confidence Interval (95%)
sample = tips['total_bill'].sample(n=30, random_state=42)
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
ci = stats.t.interval(alpha=0.95, df=29, loc=sample_mean, scale=sample_std/np.sqrt(30))
print(f"95% Confidence Interval for Total Bill: {ci}")
```
